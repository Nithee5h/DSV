```python
#Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = '/content/happiness_2017.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
# Distribution of Happiness Score

sns.histplot(df['HappinessScore'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
```

**Interpretation:**

The scores form a roughly bell-shaped curve, with most countries clustering between 5 and 6, showing a slight negative skew.

```python
#Average Happiness Score by Region

region_happiness = df.groupby('Region')['HappinessScore'].mean().reset_index()
sns.barplot(x='HappinessScore', y='Region', data=region_happiness, palette='viridis')
plt.title('Average Happiness Score by Region')
plt.xlabel('Average Happiness Score')
plt.ylabel('Region')
```

**Interpretation:**

Australia, North America, and Western Europe rank highest, while Sub-Saharan Africa and South Asia report the lowest average scores.



```python
#Correlation Heatmap

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
```

**Interpretation:**

Happiness strongly correlates with GDP, social support, and health, while corruption shows a noticeable negative relationship with happiness.


```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Happiness vs Log GDP per capita
sns.scatterplot(ax=axes[0], x='Log GDP per capita', y='HappinessScore', data=df, hue='Region', legend=False)
axes[0].set_title('Happiness Score vs Log GDP per capita')

# Happiness vs Social Support
sns.scatterplot(ax=axes[1], x='Social support', y='HappinessScore', data=df, hue='Region', legend=False)
axes[1].set_title('Happiness Score vs Social Support')

# Happiness vs Healthy life expectancy
sns.scatterplot(ax=axes[2], x='Healthy life expectancy at birth', y='HappinessScore', data=df, hue='Region', legend='brief')
axes[2].set_title('Happiness Score vs Healthy Life Expectancy')
```

**Interpretation:**

Strong positive trends exist where higher GDP, social support, and life expectancy consistently correspond to higher happiness scores globally.

```python
# Box Plot of Happiness Score by Region

sorted_regions = df.groupby('Region')['HappinessScore'].median().sort_values(ascending=False).index
sns.boxplot(x='HappinessScore', y='Region', data=df, order=sorted_regions)
plt.title('Box Plot of Happiness Score by Region')
plt.xlabel('Happiness Score')
plt.ylabel('Region')
```

**Interpretation:**

Western Europe shows consistently high happiness, whereas the Middle East and Africa display wide variations and lower median scores.
